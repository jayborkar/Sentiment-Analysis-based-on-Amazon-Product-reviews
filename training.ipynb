```python
import pandas as pd
import numpy as np
import re
from random import randint
```

```python
wordsList = np.load('wordsList.npy')
wordsList = wordsList.tolist()
wordsList = [word.decode('UTF-8') for word in wordsList] #Encode words as UTF-8
wordVectors = np.load('wordVectors.npy')

df = pd.read_csv('test.csv')
df = df.sample(int(len(df)/5))
msk = np.random.rand(len(df)) < 0.8
```

```python
len(df)
```

```python
strip_special_chars = re.compile("[^A-Za-z0-9 ]+")
def cleanSentences(string):
    string = string.lower().replace("<br />", " ")
    return re.sub(strip_special_chars, "", string.lower())
```

```python
comment_count = 0
word_count = 0
for index, row in df.iterrows():
    cleanedLine = cleanSentences(row['Text'])
    split = cleanedLine.split()
    for word in split:
        word_count += 1
    comment_count += 1
print('The average length of comments is:', word_count/comment_count)
```

```python
maxSeqLength = 100
```

```python
def getTrainBatch(msk, ids, labels):
    train_arr = ids[msk]
    train_labels = labels[msk]

    batchSize = 24

    arr = np.zeros([batchSize, maxSeqLength])
    labels = np.zeros([batchSize, 2])
    for i in range(batchSize):
        num = randint(1, len(train_arr))
        arr[i] = train_arr[num-1:num]
        labels[i] = train_labels[num-1:num]
    return arr, np.ndarray.tolist(labels)

def getTestBatch(msk, ids, labels):
    test_arr = ids[~msk]
    test_labels = labels[~msk]

    batchSize = 24

    arr = np.zeros([batchSize, maxSeqLength])
    labels = np.zeros([batchSize, 2])
    for i in range(batchSize):
        num = randint(1, len(test_arr))
        arr[i] = test_arr[num-1:num]
        labels[i] = test_labels[num-1:num]
    return arr, np.ndarray.tolist(labels)
```

```python
ids = np.zeros((len(df), maxSeqLength), dtype='int32')
commentCounter = 0

for index, row in df.iterrows():
    indexCounter = 0
    cleanedLine = cleanSentences(row['Text'])
    split = cleanedLine.split()
    sentiment = row['Sentiment']
    for word in split:
        if indexCounter < maxSeqLength:
            try:
                ids[commentCounter][indexCounter] = wordsList.index(word)
            except ValueError:
                ids[commentCounter][indexCounter] = 399999
            indexCounter += 1
            if indexCounter >= maxSeqLength:
                break
    commentCounter = commentCounter + 1
    if commentCounter % 30 == 0:
        print(commentCounter/len(df))

np.save('idsMatrix', ids)
```

```python
label = []
for index, row in df.iterrows():
    sentiment = row['Sentiment']
    if sentiment ==2:
        label.append([1,0])
    else:
        label.append([0,1])
label = np.asarray(label)
```

```python
ids = np.load('idsMatrix.npy')
```

```python
ids[25]
```

```python
batchSize = 24
lstmUnits = 64
numClasses = 2
iterations = 100000
numDimensions = 300
```

```python
import tensorflow as tf
```

```python
tf.reset_default_graph()
```

```python
labels = tf.placeholder(tf.float32, [batchSize, numClasses])
input_data = tf.placeholder(tf.int32, [batchSize, maxSeqLength])

data = tf.Variable(tf.zeros([batchSize, maxSeqLength, numDimensions]),dtype=tf.float32)
data = tf.nn.embedding_lookup(wordVectors,input_data)

lstmCell = tf.contrib.rnn.BasicLSTMCell(lstmUnits)
lstmCell = tf.contrib.rnn.DropoutWrapper(cell=lstmCell, output_keep_prob=0.75)
value, _ = tf.nn.dynamic_rnn(lstmCell, data, dtype=tf.float32)

weight = tf.Variable(tf.truncated_normal([lstmUnits, numClasses]))
bias = tf.Variable(tf.constant(0.1, shape=[numClasses]))
value = tf.transpose(value, [1, 0, 2])
last = tf.gather(value, int(value.get_shape()[0]) - 1)
prediction = (tf.matmul(last, weight) + bias)

correctPred = tf.equal(tf.argmax(prediction,1), tf.argmax(labels,1))
accuracy = tf.reduce_mean(tf.cast(correctPred, tf.float32))

loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=prediction, labels=labels))
optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss)
```

```python
import datetime

sess = tf.InteractiveSession()
tf.summary.scalar('Loss', loss)
tf.summary.scalar('Accuracy', accuracy)
merged = tf.summary.merge_all()
logdir = "tensorboard/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S") + "/"
writer = tf.summary.FileWriter(logdir, sess.graph)


saver = tf.train.Saver()
sess.run(tf.global_variables_initializer())
```

```python
# Training:

for i in range(iterations):
    # Next Batch of reviews
    nextBatch, nextBatchLabels = getTrainBatch(msk, ids, label);
    sess.run(optimizer, {input_data: nextBatch, labels: nextBatchLabels})

    # Write summary to Tensorboard
    if (i % 50 == 0):
        summary = sess.run(merged, {input_data: nextBatch, labels: nextBatchLabels})
        writer.add_summary(summary, i)

    # Save the network every 10,000 training iterations
    if (i % 10000 == 0 and i != 0):
        save_path = saver.save(sess, "models_final/pretrained_lstm.ckpt", global_step=i)
        print("saved to %s" % save_path)
writer.close()
```

```python
sess = tf.InteractiveSession()
saver = tf.train.Saver()
saver.restore(sess, tf.train.latest_checkpoint('models_final'))
```

```python
iterations = 30
avg_accuracy = 0
for i in range(iterations):
    nextBatch, nextBatchLabels = getTestBatch(msk,ids,label);
    acc = (sess.run(accuracy, {input_data: nextBatch, labels: nextBatchLabels})) * 100
    print("Accuracy for this batch:", acc)
    avg_accuracy += acc
    
avg_accuracy = avg_accuracy/iterations
print("average accuracy;", avg_accuracy)
```
